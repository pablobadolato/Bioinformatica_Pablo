#### 7.- Con la secuencia y la lista siguiente, escribe un algoritmo que cuente el número de veces que aparece un elemento de la lista en la secuencia. Pista: hacer un diccionario para almacenar los resultados puede ser buena idea

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python
repetidos = {}
for elemento in dinucleotides:
    a = dna.count(elemento)
    repetidos[elemento]= a
print(f"En tu secuencia de dna tiene los siguiente dinucleótidos repetidos {repetidos}")
```

##### ¿Cuantos counts tiene TC?

```python
repetidos["TC"]
```

##### ¿Puedes contar los dinucleótidos que tengan counts igual a 2?

```python
for elemento in repetidos:
    if repetidos[elemento] == 2:
        print(elemento)
```
