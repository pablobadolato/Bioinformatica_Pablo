```python
# Intento 3 
import random
print ("Hola jugador, bienvenido a JUEGO DE ADIVINANZAS")
print ("El juego consiste en adivinar un numero entero entre 1 y 100. Yo voy a elegir un número al azar y tú tienes que intentar adivinarlo")
print ("Si te has acercado en un rango de 10 números, te diré: '¡CALIENTE!'. A medida que te vayas acercando más te iré diciendo '¡MAS CALIENTE!' hasta que lo adivines. Sin embargo, si te alejas te diré '¡MAS FRIO!'")
def pedir_numero():
    #Hago este bucle porque sino me da error al meter un valor no numérico. Es decir, si respondo con un "hola" se ejecuta el exept y al pasar al if, la variable "respuesta" no está definida, de esta forma fuerzo a definir "respuesta" para luego poder llamarla en el if
    while True:
        try:
            respuesta = int(input("Escoge un número entero entre 1 y 100"))
        except ValueError:
            print("No has introducido un número entero")
            #este continue permite pasar al if habiendo definido respuesta como un número entero
            continue
        if not (1 <= respuesta <= 100):
            print("Fuera de limites")
            respuesta = int(input("Fuera de limites. Escoge un número entero entre 1 y 100"))
        return respuesta
def juego():
    #Defino cual es el número aleatorio
    numero = random.randint(1, 101)
    #Establezco el rango de ¡CALIENTE!, que es número +-10. Pongo max y min para que el rango esté siempre entre 1 y 100
    rango_numero_alto = min(100, numero + 10) #elige el número menor (así no pasa de 100)
    rango_numero_bajo = max(1, numero - 10) #elige el número mayor (así no pasa de 1)
    #Esto lo vas a usar para mirar si es mayor o menor tu último guess. Para el +/- 10 que hay que ir poniendo +caliente o +frio. Uso None porque me permite calcular más facilmente como un int en vez de "" que es una str y me da muchos problemas
    ultimo = None
    #Creo una lista donde se guarden todos los intentos del jugador
    intentos = []
    #Hago un bucle para permitir varios intentos hasta descubrir el valor de numero
    while True:
         #Incluyo la función pedir_numero con la que voy a conseguir los intentos
        intento = pedir_numero()
        #Añado todos los intentos hechos a la lista de intentos
        intentos.append(intento)
        #Comienzo con la adivinación
        if intento == numero:
            print(f"¡Lo conseguiste! Has necesitado {len(intentos)} intentos para adivinar el número. ¡Enhorabuena!")
            break
        #Compruebo si está en el rango de ¡CALIENTE!
        elif intento in range (rango_numero_bajo, rango_numero_alto + 1):
            #Una vez se haya visto que está en caliente, compruebo si está más cerca o más lejos que el anterior
            #Primero veo si el último intento estaba o no en ¡CALIENTE!, si no lo está sólo hay que decir que ha entrado en ¡CALIENTE!
            if ultimo is None or not (rango_numero_bajo <= ultimo <= rango_numero_alto):
                print("¡CALIENTE!")
            #Si el jugador ya estaba en ¡CALIENTE!, debo comprobar si se acerca más al resultado o menos
            else:
                distancia_actual = abs(numero - intento)
                distancia_anterior = abs(numero - ultimo)
                if distancia_actual < distancia_anterior:
                    print("¡MAS CALIENTE!")
                elif distancia_actual > distancia_anterior:
                    print("¡MAS FRIO!")
                else:
                    print("Igual de caliente")
            #Asigno el último intento como ultimo para poder calcular si se acerca más o menos al numero. Lo hago al final para que me registre una vez se haya corrido todo el código
            ultimo = intento
        #Si no está en el rango de caliente, digo que está frio
        elif not (rango_numero_bajo <= intento <= rango_numero_alto):
            print ("¡FRIO!")
juego()
```
