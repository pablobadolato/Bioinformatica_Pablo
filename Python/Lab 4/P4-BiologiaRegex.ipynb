##### ¿Qué pasa si no ponemos un argumento? ¿Sabrías hacer una función encapsulada?

```python
#Si no pones ningún argumento, me va a dar error porque no sabe a qué me refiero. Decimales lo tengo establecido como un valor por defecto, por lo que por eso no me va a dar error, pero dna no está definido y sí me dará un error. Si dna no está definido en la función, buscará el valor fuera de la función, si da la casualidad de que tienes llamado así a tu secuencia no habrá problema, pero si dna es una lista o no está definida o cualquier cosa del estilo, me dará un error. El valor de dna lo puede buscar en otras funciones y, si es así, sería una función encapsulada
#ejemplo
def analizar():
    # variable definida en la función principal
    dna = "AAGTCTTA"

    # función interna (encapsulada)
    def porcentaje_A(decimales=2):
        # usa dna que está definida arriba (en la función que la contiene)
        dna_mayus = dna.upper()
        tamaño_adn = len(dna_mayus)
        tamaño_A = dna_mayus.count("A")
        porcentajeA = (tamaño_A / tamaño_adn) * 100
        return round(porcentajeA, decimales)
    
    # la función interna solo se puede usar aquí
    return porcentaje_A()

analizar()
```

#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re
```

```python
accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']
```

```python
#Los que tengan el número 5
con5 = []
for accession in accs:
    if re.search("5", accession):
        con5.append(accession)
con5
```

```python
#Los que tengan la letra 'd' o la 'e'
conDoE = []
for accession in accs:
    if re.search("(d|e)", accession):
        conDoE.append(accession)
conDoE
```

```python
#Los que tengan las letras 'd' y 'e' en ese orden
conDyE = []
for accession in accs:
    if re.search("de", accession):
        conDyE.append(accession)
conDyE
```

```python
#Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
conDyEsinnum = []
for accession in accs:
    if re.search(r"d.e", accession):
        conDyEsinnum.append(accession)
conDyEsinnum
```

```python
#Los que tengan ambas letras 'd' y 'e' en cualquier orden
conDyE = []
for accession in accs:
    if re.search("d", accession) and re.search("e", accession):
        conDyE.append(accession)
conDyE
```

```python
#Los que empiezan por 'x' o por 'y'
porXoY = []
for accession in accs:
    if re.search(r"^(x|y)", accession):
        porXoY.append(accession)
porXoY
```

```python
#Los que empiezan por 'x' o 'y' y terminan por 'e'
XoYfinE = []
for accession in accs:
    if re.search(r"^(x|y).*e$", accession):
        XoYfinE.append(accession)
XoYfinE
```

```python
#Los que tienen tres o más dígitos seguidos
tresomas = []
for accession in accs:
    if re.search(r"[0-9]{3}", accession):
        tresomas.append(accession)
tresomas
```

```python
#Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'
terminan = []
for accession in accs:
    if re.search(r"d(a|r|p)$", accession):
        terminan.append(accession)
terminan
```
